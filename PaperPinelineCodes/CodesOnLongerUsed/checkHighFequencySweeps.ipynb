```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
direc = r"../dataFolders/PaperPipelineOutput/FilteredTracks_v2/"
files = glob.glob(direc + 'c*.csv')
```

```python
files[1]
```

```python
file = [f for f in files if 'c-10_m23_' in f]

for f in file:
    trajectory = pd.read_csv(f)
    name = os.path.basename(f)[:-15]
    plt.plot(np.arange(len(trajectory.r))*1/100,
             trajectory.r, label = name)
    plt.legend()
    plt.xlim(10, 20)
    plt.ylim(-0.2, 5)
```
